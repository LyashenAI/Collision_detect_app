```python
from tkinter import *
from tkinter import scrolledtext
from tkinter import filedialog
from tkinter import Tk, Canvas, Frame, BOTH
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageTk
import cv2
import imutils

def clear():
    txt_pole.delete(1.0, END)
    txt_pole1.delete(1.0, END)
def show():
    global x1
    global x2
    global x3
    global x4
    global x5
    x1 = var1.get()
    x2 = var2.get()
    x3 = var3.get()
    x4 = var4.get()
    x5 = var5.get()
    if var4.get()==1:
        s='Нажата - Сети связи'
        txt_pole1.insert(END, s)
        txt_pole1.insert("999.0", "\n")
        x1=1
    elif var5.get()==1:
        s='Нажата - Наружное водоснабжение'
        txt_pole1.insert(END, s)
        txt_pole1.insert("999.0", "\n")
        x2=1
    elif var6.get()==1:
        s='Нажата - Дождевая канализация'
        txt_pole1.insert(END, s)
        txt_pole1.insert("999.0", "\n")
        x3=1
    elif var7.get()==1:
        s='Нажата - Электрические сети'
        txt_pole1.insert(END, s)
        txt_pole1.insert("999.0", "\n")
        x4=1
    elif var8.get()==1:
        s='Нажата - Канализация'
        txt_pole1.insert(END, s)
        txt_pole1.insert("999.0", "\n")
        x5=1
def open_file():
        file = filedialog.askopenfilename(initialdir='/', title='Open file', filetypes=(('Text Documents', '*.txt'), ('all files', '*.*')))
        if file: 
            f = open(file, 'r')
            text_open = f.read()
            if text_open != NONE:
                txt_pole.delete(1.0, END)
                txt_pole.insert(END, text_open)
def izo():
    rgb_img = cv2.imread('1.png')
    gray_img = cv2.cvtColor(rgb_img, cv2.COLOR_BGR2GRAY)
    cv2.imshow('gray image', imutils.resize(gray_img, 800))
    cv2.imwrite('gray_img.png', gray_img)
    if cv2.waitKey(0) == 27:
        cv2.destroyAllWindows()
def save_file():
    file = filedialog.asksaveasfilename(initialdir='/', title='Select file', filetypes=(('Text Documents', '*.txt'), ('all files', '*.*')))
    if file:
        f = open(file, 'w')
        text_save = str(txt_pole1.get(1.0, END))
        f.write(text_save+'\n') 
        f.close()
        
    def __init__(self, parent):
        Frame.__init__(self, parent)
        self.parent = parent
        self.color = "red"
        self.brush_size = 2
        self.setUI()
    def draw(self, event):
        self.canv.create_oval(event.x - 2, event.y - 2, event.x + 2, event.y + 2, fill=self.color, outline=self.color)
    def setUI(self):
        self.pack(fill=BOTH, expand=1)
        self.canv = Canvas(self, bg="white")
        self.columnconfigure(0, weight=1)
        self.rowconfigure(0, weight=1)
        self.canv.grid(padx=5, pady=5, sticky=E + W + S + N)
        self.canv.bind("<B1-Motion>", self.draw)
def vibor():
    root = Tk()
    if var4.get() == 1:
        app = Paint(root)
    elif var5.get() == 1:
        app = Paint(root)        
    elif var6.get() == 1:
        app = Paint(root)
    elif var7.get() == 1:
        app = Paint(root)
    elif var8.get() == 1:
        app = Paint(root)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

class Neuron:
    def __init__(self, weights, bias):
        self.weights = weights
        self.bias = bias 
 
    def feedforward(self, inputs): 
        total = np.dot(self.weights, inputs) + self.bias
        return sigmoid(total)
    
class OurNeuralNetwork:
    def __init__(self):
        weights = np.array([1, 1, 2, 2, 2])
        bias = 0         
        self.h1 = Neuron(weights, bias)         
        self.h2 = Neuron(weights, bias)         
        self.h3 = Neuron(weights, bias)         
        self.h4 = Neuron(weights, bias)         
        self.h5 = Neuron(weights, bias)         
        self.o1 = Neuron(weights, bias) 
 
    def feedforward(self, x): 
        out_h1 = self.h1.feedforward(x)         
        out_h2 = self.h2.feedforward(x)         
        out_h3 = self.h2.feedforward(x)         
        out_h4 = self.h2.feedforward(x)         
        out_h5 = self.h2.feedforward(x)         
        out_o1 = self.o1.feedforward(np.array([out_h1, out_h2, out_h3, out_h4, out_h5])) 
 
        return out_o1 
 
def ss():
    weights = np.array([1,1,2,2,2])
    network = OurNeuralNetwork() 
    x = np.array([x1,x2,x3,x4,x5])  
    metka="Значение нейрона"     
    txt_pole1.insert(END, metka)     
    txt_pole1.insert("999.0","\n") 
 
    s=network.feedforward(x) 
 
    txt_pole1.insert(END, s)     
    txt_pole1.insert("999.0","\n")     
    if s<0.5:
        m1="Коллизий нет"         
        txt_pole1.insert(END, m1)         
        txt_pole1.insert("999.0","\n")     
    else: 
        m2="Необходимо переделать"         
        txt_pole1.insert(END, m2)         
        txt_pole1.insert("999.0","\n") 

window = Tk() 
window.title("Разработка ПМ поиска коллизий") 
window.geometry('700x800') 
txt_pole = scrolledtext.ScrolledText(window, width=60, height=20)   
txt_pole.grid(column=0, row=0, columnspan=5, rowspan=5)  
lbl = Label(window, text="Управление", font=("Arial Bold", 12))   
lbl.grid(column=5, row=0)   
btn1 = Button(window, text="Загрузить",bg="#96d3ff", fg="#0e047f", command=open_file)   
btn1.grid(column=5, row=1)
btn1 = Button(window, text="Сохранить",bg="#96d3ff", fg="#0e047f", command=save_file)   
btn1.grid(column=5, row=2)
btn2 = Button(window, text="Очистить",bg="#96d3ff", fg="#0e047f", command=clear)   
btn2.grid(column=5, row=3)
btn2 = Button(window, text="Рассчитать",bg="#96d3ff", fg="#0e047f", command=ss)   
btn2.grid(column=5, row=4)

var1 = BooleanVar() 
var1.set(0) 
c1 = Checkbutton(window, text="Сеть связи", variable=var1, onvalue=1, offvalue=0, command=show) 
c1.grid(column=0, row=5) 
var2 = BooleanVar() 
var2.set(0) 
c2 = Checkbutton(window, text="Наружное водоснабжение", variable=var2, onvalue=1, offvalue=0, command=show) 
c2.grid(column=1, row=5) 
var3 = BooleanVar() 
var3.set(0) 
c3 = Checkbutton(window, text="Дождевая канализация", variable=var3, onvalue=1, offvalue=0, command=show) 
c3.grid(column=2, row=5) 
var4 = BooleanVar() 
var4.set(0) 
c4 = Checkbutton(window, text="Электрические сети", variable=var4, onvalue=1, offvalue=0, command=show) 
c4.grid(column=3, row=5) 
var5 = BooleanVar() 
var5.set(0) 
c5 = Checkbutton(window, text="Канализация", variable=var5, onvalue=1, offvalue=0, command=show) 
c5.grid(column=0, row=6) 

txt_pole1 = scrolledtext.ScrolledText(window, width=60, height=20)   
txt_pole1.grid(column=0, row=7, columnspan=5, rowspan=5)  

btn3 = Button(window, text="Изображение",bg="#96d3ff", fg="#0e047f", command=izo)   
btn3.grid(column=5, row=8) 
btn4 = Button(window, text="  Добавить ",bg="#96d3ff", fg="#0e047f", command=vibor)   
btn4.grid(column=5, row=9) 
window.mainloop() 



  
    




        
```
